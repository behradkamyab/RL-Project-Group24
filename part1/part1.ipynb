```python
import pickle
import os
import numpy as np
import matplotlib.pyplot as plt

def load_time(suffix):
    path = os.path.join(PREFIX, f'time_{suffix}.pkl')
    with open(path, 'rb') as f:
        return pickle.load(f)

PREFIX = 'results'

def smooth(x, w=20):
    return np.convolve(x, np.ones(w)/w, mode='valid')

def smooth_50(x, w=50):
    return np.convolve(x, np.ones(w)/w, mode='valid')

def load_history(suffix):
    path = os.path.join(PREFIX, f'history_{suffix}.pkl')
    with open(path, 'rb') as f:
        return pickle.load(f)

def final_mean(suffix, last_n=50):
    h = load_history(suffix)
    return float(np.mean(h[-last_n:]))

EMA_SLOW   = 'ema_a0.0005'
EMA_MEDIUM = 'ema_a0.0500'
EMA_FAST   = 'ema_a0.5000'
```

```python
fig, (ax_const, ax_ema) = plt.subplots(1, 2, figsize=(14, 5))

# --- Left: constant baseline sensitivity ---
const_runs = [
    ('none',          'vanilla (no preproc)',      {'linestyle': ':',  'color': 'black'}),
    ('constant_b0',   'b = 0',                     {}),
    ('constant_b20',  'b = 20',                    {}),
    ('constant_b100', 'b = 100',                   {}),
]
for suffix, label, style in const_runs:
    h = load_history(suffix)

    ax_const.plot(smooth(h), label=label, **style)

ax_const.set_xlabel('Episode')
ax_const.set_ylabel('Return (20-ep moving avg)')
ax_const.set_title('(a) Constant baseline sensitivity')
ax_const.legend(fontsize=9)
ax_const.grid(alpha=0.3)

# --- Right: EMA alpha sensitivity ---
ema_runs = [
    ('none',     'vanilla (no preproc)',      {'linestyle': ':',  'color': 'black'}),
    (EMA_SLOW,   'α = 0.0005',                {}),
    (EMA_MEDIUM, 'α = 0.05',                  {}),
    (EMA_FAST,   'α = 0.5',                   {}),
]
for suffix, label, style in ema_runs:
    h = load_history(suffix)
    ax_ema.plot(smooth(h), label=label, **style)

ax_ema.set_xlabel('Episode')
ax_ema.set_ylabel('Return (20-ep moving avg)')
ax_ema.set_title('(b) EMA α sensitivity')
ax_ema.legend(fontsize=9)
ax_ema.grid(alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(PREFIX, 'sensitivity_analysis.png'),
            dpi=120, bbox_inches='tight')
plt.show()
print("\nSaved: sensitivity_analysis.png")
```

## Refining Actor-Critic from the sweep winner

`ac_trial_7` peaks at ~621 around episode 1364, then **decays to ~400** by episode 3000 — the failure mode is *late-training instability*, not under-training. So the goal here is to **hold the peak**, not train longer.

These runs start from ac_7's recipe (low `actor_lr`, fast critic, `hidden=256`, `gamma=0.995`) and add stabilizers — `normalize_advantage=True`, a lower `actor_lr`, and a tighter `grad_clip` — across multiple seeds. We rank by **peak** (best-case) and by the **peak-to-final gap** (smaller = more stable).

```python
refined_cfg = dict(
    actor_lr=2.5e-5,          
    critic_lr=4.0e-4,         
    critic_weight=1.0,
    hidden_size=256,
    gamma=0.995,
    grad_clip=0.3,            
    entropy_weight=1e-4,
    normalize_advantage=True, 
)

REFINE_SEEDS = [0, 1, 2, 3, 4, 5, 6, 7, 42]

for seed in REFINE_SEEDS:
    name = f"ac_refined_seed{seed}"
    print(f"\n=== {name} ===")
    main(algorithm="actor_critic", n_episodes=3000, seed=seed,
         run_name=name, save_dir=PREFIX, **refined_cfg)
```

```python
RUN_SEARCH = True
SEARCH_SEEDS = REFINE_SEEDS          

search_grid = [
    {"actor_lr": 3.5e-5, "normalize_advantage": True},
    {"actor_lr": 2.5e-5, "normalize_advantage": False},
    {"actor_lr": 3.5e-5, "normalize_advantage": False},
]

if RUN_SEARCH:
    for i, override in enumerate(search_grid):
        cfg = {**refined_cfg, **override}
        for seed in SEARCH_SEEDS:
            name = f"ac_search_{i}_seed{seed}"
            print(f"\n=== {name}: {override} ===")
            main(algorithm="actor_critic", n_episodes=3000, seed=seed,
                 run_name=name, save_dir=PREFIX, **cfg)
```

## Promote the refined Actor-Critic into the multi-seed & fair comparison

These cells pick the winning refined config (by peak), evaluate it over seeds `[0, 1, 2]` like the original `actor_critic` (cell 19), and add it as an extra line to the fair 3000-episode comparison (cell 20). The existing cells are left untouched.

```python
import os, json as _json

ac_candidates = {f"ac_refined_seed{s}": dict(refined_cfg) for s in REFINE_SEEDS}
if RUN_SEARCH:
    for i, override in enumerate(search_grid):
        cfg = {**refined_cfg, **override}
        for s in SEARCH_SEEDS:
            ac_candidates[f"ac_search_{i}_seed{s}"] = cfg

def _exists(suffix):
    return os.path.exists(os.path.join(PREFIX, f"history_{suffix}.pkl"))

def _peak_gap(suffix, last_n=50):
    h = load_history(suffix)
    sm = smooth_50(h)
    peak = float(sm.max())
    final = float(np.mean(h[-last_n:]))
    return peak, peak - final

available = {n: c for n, c in ac_candidates.items() if _exists(n)}
if not available:
    raise RuntimeError("Run the refined cells (16-18) first to generate histories.")

groups = {}  
for name, cfg in available.items():
    g = groups.setdefault(_json.dumps(cfg, sort_keys=True), {"cfg": cfg, "runs": []})
    g["runs"].append(name)

ranking = [] 
for g in groups.values():
    stats = [_peak_gap(n) for n in g["runs"]]
    mean_peak = float(np.mean([s[0] for s in stats]))
    mean_gap = float(np.mean([s[1] for s in stats]))
    rep = max(g["runs"], key=lambda n: _peak_gap(n)[0])  
    ranking.append((mean_peak, mean_gap, g["cfg"], g["runs"], rep))

ranking.sort(key=lambda r: (-r[0], r[1]))

REFINED_BEST_NAME = None

print(f"{'mean_peak':>10}{'mean_gap':>10}{'n':>3}  config")
for i, (mean_peak, mean_gap, cfg, runs, rep) in enumerate(ranking):
    tag = "  <-- best" if i == 0 else ""
    print(f"{mean_peak:10.0f}{mean_gap:10.0f}{len(runs):3d}  "
          f"alr={cfg['actor_lr']:.1e} clr={cfg['critic_lr']:.1e} "
          f"normAdv={cfg['normalize_advantage']} gc={cfg['grad_clip']}{tag}")

best_mean_peak, best_mean_gap, REFINED_BEST_CFG, _, best_rep = ranking[0]
if REFINED_BEST_NAME is None:
    REFINED_BEST_NAME = best_rep   

print(f"\nPromoting config (mean_peak={best_mean_peak:.0f}, "
      f"mean_gap={best_mean_gap:.0f}) via run '{REFINED_BEST_NAME}'")
print("Config:", REFINED_BEST_CFG)
```

Checks whether bootstrapping through the 1000-step time limit (vs. treating it as a
terminal state) affects AC. Trains both variants on matched seeds; justifies the
(1 - terminated) choice in agent.py.

```python

import gymnasium as gym
import torch, random, pickle

from agent import Agent, Policy

def _ac_cfg():
    """Best AC config, normalized to hidden_size; prefer the refined one if present."""
    try:
        return dict(REFINED_BEST_CFG)
    except NameError:
        return dict(actor_lr=best_cfg["actor_lr"], critic_lr=best_cfg["critic_lr"],
                    critic_weight=best_cfg["critic_weight"], hidden_size=best_cfg["hidden"],
                    gamma=best_cfg["gamma"], grad_clip=best_cfg["grad_clip"],
                    entropy_weight=best_cfg["entropy_weight"],
                    normalize_advantage=best_cfg["normalize_advantage"])

def train_ac(bootstrap_truncation, cfg, seed, n_episodes=3000, run_name=None):
    torch.manual_seed(seed); np.random.seed(seed); random.seed(seed)
    env = gym.make("Hopper-v4")
    env.reset(seed=seed); env.action_space.seed(seed)
    policy = Policy(state_space=env.observation_space.shape[0],
                    action_space=env.action_space.shape[0], hidden=cfg["hidden_size"])
    agent = Agent(policy, algorithm="actor_critic",
                  actor_lr=cfg["actor_lr"], critic_lr=cfg["critic_lr"],
                  critic_weight=cfg["critic_weight"], gamma=cfg["gamma"],
                  grad_clip=cfg["grad_clip"], entropy_weight=cfg["entropy_weight"],
                  normalize_advantage=cfg["normalize_advantage"])
    history = []
    for ep in range(n_episodes):
        state, _ = env.reset(); done = False; ep_rew = []
        while not done:
            action, log_prob, entropy = agent.get_action(state)
            action = action.detach().cpu().numpy()
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            ep_rew.append(reward)

            if bootstrap_truncation:
                term, trunc = terminated, truncated
            else:
                term, trunc = (terminated or truncated), False
            agent.store_outcome(state, next_state, log_prob, entropy, reward, term, trunc)
            state = next_state
        agent.update_policy()
        history.append(sum(ep_rew))
        if ep % 100 == 0:
            print(f"  ep {ep:4d}  return={sum(ep_rew):5.0f}")
    env.close()
    if run_name is not None:
        with open(os.path.join(PREFIX, f"history_{run_name}.pkl"), "wb") as f:
            pickle.dump(history, f)
    return history

try:
    TRUNC_SEEDS = CMP_SEEDS              
except NameError:
    TRUNC_SEEDS = [0, 1, 2]

trunc_variants = [
    ("ac_trunc_terminal",  False),      
    ("ac_trunc_bootstrap", True),        
]

ac_cfg = _ac_cfg()
print("AC config under test:", ac_cfg)
for prefix, boot in trunc_variants:
    for seed in TRUNC_SEEDS:
        run_name = f"{prefix}_seed{seed}"
        if os.path.exists(os.path.join(PREFIX, f"history_{run_name}.pkl")):
            print("skip (cached):", run_name); continue
        print(f"running: {run_name}  (bootstrap_truncation={boot})")
        train_ac(boot, ac_cfg, seed, run_name=run_name)

print("\nDone:", [f"{p}_seed*" for p, _ in trunc_variants], "| seeds:", TRUNC_SEEDS)
```
