```python
import pickle
import os
import numpy as np
import matplotlib.pyplot as plt

def load_time(suffix):
    path = os.path.join(PREFIX, f'time_{suffix}.pkl')
    with open(path, 'rb') as f:
        return pickle.load(f)

PREFIX = 'results'

def smooth(x, w=20):
    return np.convolve(x, np.ones(w)/w, mode='valid')

def smooth_50(x, w=50):
    return np.convolve(x, np.ones(w)/w, mode='valid')

def load_history(suffix):
    path = os.path.join(PREFIX, f'history_{suffix}.pkl')
    with open(path, 'rb') as f:
        return pickle.load(f)

def final_mean(suffix, last_n=50):
    h = load_history(suffix)
    return float(np.mean(h[-last_n:]))

EMA_SLOW   = 'ema_a0.0005'
EMA_MEDIUM = 'ema_a0.0500'
EMA_FAST   = 'ema_a0.5000'
```

```python
fig, (ax_const, ax_ema) = plt.subplots(1, 2, figsize=(14, 5))

# --- Left: constant baseline sensitivity ---
const_runs = [
    ('none',          'vanilla (no preproc)',      {'linestyle': ':',  'color': 'black'}),
    ('normalize',     'per-ep normalization',      {'linestyle': '--', 'color': 'gray'}),
    ('constant_b0',   'b = 0',                     {}),
    ('constant_b20',  'b = 20',                    {}),
    ('constant_b100', 'b = 100',                   {}),
]
for suffix, label, style in const_runs:
    h = load_history(suffix)

    ax_const.plot(smooth(h), label=label, **style)

ax_const.set_xlabel('Episode')
ax_const.set_ylabel('Return (20-ep moving avg)')
ax_const.set_title('(a) Constant baseline sensitivity')
ax_const.legend(fontsize=9)
ax_const.grid(alpha=0.3)

# --- Right: EMA alpha sensitivity ---
ema_runs = [
    ('none',     'vanilla (no preproc)',      {'linestyle': ':',  'color': 'black'}),
    ('normalize','per-ep normalization',      {'linestyle': '--', 'color': 'gray'}),
    (EMA_SLOW,   'α = 0.0005',                {}),
    (EMA_MEDIUM, 'α = 0.05',                  {}),
    (EMA_FAST,   'α = 0.5',                   {}),
]
for suffix, label, style in ema_runs:
    h = load_history(suffix)
    ax_ema.plot(smooth(h), label=label, **style)

ax_ema.set_xlabel('Episode')
ax_ema.set_ylabel('Return (20-ep moving avg)')
ax_ema.set_title('(b) EMA α sensitivity')
ax_ema.legend(fontsize=9)
ax_ema.grid(alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(PREFIX, 'sensitivity_analysis.png'),
            dpi=120, bbox_inches='tight')
plt.show()
print("\nSaved: sensitivity_analysis.png")
```

```python
def final_mean(suffix, last_n=50):
    """Average return over the last N episodes of a run."""
    h = load_history(suffix)
    return float(np.mean(h[-last_n:]))

# Constant variants
const_candidates = ['constant_b0', 'constant_b20', 'constant_b100']
const_scores = {c: final_mean(c) for c in const_candidates}
BEST_CONST = max(const_scores, key=const_scores.get)
print("Constant scores:", const_scores)
print(f"Best constant: {BEST_CONST}")

# EMA variants
ema_candidates = [EMA_SLOW, EMA_MEDIUM, EMA_FAST]
ema_scores = {e: final_mean(e) for e in ema_candidates}
BEST_EMA = max(ema_scores, key=ema_scores.get)
print("EMA scores:", ema_scores)
print(f"Best EMA: {BEST_EMA}")

summary_runs = [
    ('none',      'vanilla (no preproc)',   {'linestyle': ':',  'color': 'black'}),
    ('normalize', 'per-ep normalization',   {'linestyle': '--', 'color': 'gray'}),
    (BEST_CONST,  'constant (best)',        {}),
    (BEST_EMA,    'EMA (best)',             {}),
]

plt.figure(figsize=(10, 5))

for suffix, label, style in summary_runs:
    h = load_history(suffix)
    plt.plot(smooth(h), label=label, **style)

plt.xlabel('Episode')
plt.ylabel('Return (20-ep moving avg)')
plt.title('REINFORCE variants on Hopper — best-performing baselines')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(os.path.join(PREFIX, 'reinforce_best_summary.png'),
            dpi=120, bbox_inches='tight')
plt.show()


print("\nDone. Saved:")
print(f"  - sensitivity_analysis.png")
print(f"  - reinforce_best_summary.png")
```

```python
import random
PREFIX = "results"
os.makedirs(PREFIX, exist_ok=True)

def loguniform(low, high):
    return float(np.exp(np.random.uniform(np.log(low), np.log(high))))

np.random.seed(123)
random.seed(123)

N_TRIALS = 12

ac_configs = []

for i in range(N_TRIALS):
    cfg = {
        "run_name": f"ac_trial_{i}",
        "actor_lr": loguniform(1e-5, 3e-4),
        "critic_lr": loguniform(3e-5, 1e-3),
        "entropy_weight": random.choice([0.0, 1e-4, 1e-3, 5e-3]),
        "critic_weight": random.choice([0.25, 0.5, 1.0]),
        "hidden": random.choice([64, 128, 256]),
        "gamma": random.choice([0.99, 0.995]),
        "grad_clip": random.choice([0.5, 1.0]),
        "normalize_advantage": random.choice([True, False]),
    }
    ac_configs.append(cfg)

for cfg in ac_configs:
    print(cfg)
```

```python
for cfg in ac_configs:
    print("\nRunning:", cfg["run_name"])
    print(cfg)

    main(
        algorithm="actor_critic",
        n_episodes=3000,
        seed=42,
        actor_lr=cfg["actor_lr"],
        critic_lr=cfg["critic_lr"],
        entropy_weight=cfg["entropy_weight"],
        critic_weight=cfg["critic_weight"],
        hidden_size=cfg["hidden"],
        gamma=cfg["gamma"],
        grad_clip=cfg["grad_clip"],
        normalize_advantage=cfg["normalize_advantage"],
        run_name=cfg["run_name"],
        save_dir=PREFIX
    )
```

```python
def ac_final_mean(run_name, last_n=20):
    h = load_history(run_name)
    return float(np.mean(h[-last_n:]))

ac_scores = {}

for cfg in ac_configs:
    name = cfg["run_name"]
    ac_scores[name] = ac_final_mean(name, last_n=20)

print("Actor-Critic sweep scores:")
for name, score in ac_scores.items():
    print(f"{name:15s} {score:.2f}")

BEST_AC = max(ac_scores, key=ac_scores.get)
print("\nBest Actor-Critic:", BEST_AC)

best_cfg = None
for cfg in ac_configs:
    if cfg["run_name"] == BEST_AC:
        best_cfg = cfg

print("\nBest config:")
print(best_cfg)
```

```python
MAX_EPISODES = 3000

BEST_AC_LABEL = (
    f"actor_lr={best_cfg['actor_lr']:.1e}, "
    f"critic_lr={best_cfg['critic_lr']:.1e}, "
    f"cw={best_cfg['critic_weight']}, "
    f"h={best_cfg['hidden']}"
)

summary_runs = [
    ('none',         'REINFORCE (vanilla)',        {'linestyle': ':',  'color': 'black'}),
    ('normalize',    'REINFORCE (per-ep norm)',    {'linestyle': '--', 'color': 'gray'}),
    (BEST_CONST,     'REINFORCE (constant, best)', {'color': 'tab:orange'}),
    (BEST_EMA,       'REINFORCE (EMA, best)',      {'color': 'tab:green'}),
    ('actor_critic', f'Actor-Critic tuned ({BEST_AC_LABEL})', {'color': 'tab:red'}),
]

print(f"Final mean return over episodes {MAX_EPISODES-50+1}-{MAX_EPISODES}:")

plt.figure(figsize=(11, 5.5))

for suffix, label, style in summary_runs:
    h = load_history(suffix)

    h_cut = h[:MAX_EPISODES]
    score = float(np.mean(h_cut[-50:]))

    print(f"  {label:70s} {score:.2f}")

    y = smooth_50(h_cut)
    x = np.arange(50, len(h_cut) + 1)

    plt.plot(
        x,
        y,
        label=f'{label}  (final@3000={score:.1f})',
        **style
    )

plt.xlabel('Episode')
plt.ylabel('Return (50-episode moving average)')
plt.title('Fair comparison: first 3000 episodes only')
plt.legend(loc='best', fontsize=8)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(os.path.join(PREFIX, 'fair_3000_episode_comparison.png'),
            dpi=120,
            bbox_inches='tight')
plt.show()

print("Saved: fair_3000_episode_comparison.png")
```
